```python
from sklearn import datasets #importing the database
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
iris = datasets.load_iris()
iris = pd.DataFrame(data= np.c_[iris['data'], iris['target']], #converting dataframe
columns= iris['feature_names'] + ['species'])  #creating labels
iris.species = np.where(iris.species == 0.0, 'setosa', np.where(iris.species==1.0,'versicolor', 'virginica'))
iris.columns = iris.columns.str.replace('','')
iris.describe()
print(iris['species'].value_counts())
```

```python
iris.hist() #Pandas dataframe visualization
plt.suptitle("Histogram", fontsize=16)
plt.show()
iris.boxplot()
plt.title("Bar Plot", fontsize=16)
plt.show()
```

```python
print(iris.groupby(by = "species").mean()) #multivariate analysis

iris.groupby(by = "species").mean().plot(kind="bar")
plt.title('Class vs Measurements')
plt.ylabel('mean measurement(cm)')
plt.xticks(rotation=0)
plt.grid(True)

plt.legend(loc="upper left", bbox_to_anchor=(1,1))
```

```python
selected_features = ['sepal.length', 'sepal.width', 'petal.length', 'petal.width']

feature_matrix = np.array(iris[selected_features]).T

mean_vector = np.sum(feature_matrix, axis=1) / feature_matrix.shape[1]

covariance_matrix = np.zeros((len(selected_features), len(selected_features)))  # covariance matrix

for i in range(len(selected_features)):
    for j in range(len(selected_features)):
        covariance_matrix[i, j] = np.sum((feature_matrix[i] - mean_vector[i]) * (feature_matrix[j] - mean_vector[j])) / (feature_matrix.shape[1] - 1)

std_dev_vector = np.sqrt(np.sum((feature_matrix - mean_vector[:, np.newaxis])**2, axis=1) / (feature_matrix.shape[1] - 1))  # Standard deviation

correlation_matrix = np.zeros((len(selected_features), len(selected_features))) # correlation matrix

for i in range(len(selected_features)):
    for j in range(len(selected_features)):
        correlation_matrix[i, j] = covariance_matrix[i, j] / (std_dev_vector[i] * std_dev_vector[j])

print(f"\nCovariance Matrix (MxM): \n{covariance_matrix}")
print(f"\nCorrelation Matrix (MxM): \n{correlation_matrix}")
```
